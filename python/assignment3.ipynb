```python
def apply_discount(price,discount_percent=5):
    if discount_percent > 60:
        raise ValueError("Discount percent must be less than 60")
    
    discount = (discount_percent*price)/100
    return price - discount

print(apply_discount(1000,10))
print(apply_discount(1000))
print(apply_discount(1000,90))
```

```python
def fac(n):
    if n == 0 or n==1:
        return n
    return fac(n-1)+fac(n-2)

fac(5)
```

```python
fac(10)
```

```python
def fact(n):
    if n<0:
        return "negetive number"
    if 1==n or 0==n:
        return n
    return n*fact(n-1)

print(fact(5))
```

```python
fact(-3)
```
